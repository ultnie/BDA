```python
#Хватит страдать фигнёй, пора подключать NLTK

import pandas as pd
import numpy as np
from datetime import datetime, timedelta, date

pd.set_option('display.max_colwidth', None)

df_x_train = pd.read_csv('./nsu-bda-2023-jobs/X_train.csv')
df_y_train = pd.read_csv('./nsu-bda-2023-jobs/y_train.csv')

df_full = pd.merge(df_x_train, df_y_train, on='id', how='outer')

import matplotlib.pyplot as plt

%matplotlib inline

df_full['employer_name'] = df_full['employer_name'].str.replace('ё', 'е').str.lower()
df_full['name'] = df_full['name'].str.replace('ё', 'е').str.lower()
df_full['area_name'] = df_full['area_name'].str.replace('ё', 'е').str.lower()

df_red = df_full.copy()
df_red.drop(['salary_currency', 'published_at', 'created_at'], axis=1, inplace=True)

df_red['salary_from'].fillna(df_full['salary_from'].mean(), inplace = True)
df_red['salary_from'].replace(to_replace=0.0, value=df_full['salary_from'].mean(), inplace = True)

binary_feats = ['salary_gross']
mapping = {'True':True, 'False':False, 'nan':False, np.nan:False, True:True, False:False}
df_red.loc[:, binary_feats] = df_red.loc[:, binary_feats].applymap(lambda x: mapping[x])

from bs4 import BeautifulSoup

df_red['description'].fillna('Empty', inplace=True)

for i in range(len(df_red['description'])):
# kill all script and style elements
    soup = BeautifulSoup(df_red.iloc[i]['description'], features="html.parser")
    for script in soup(["script", "style"]):
        script.extract()    # rip it out
#get text
    text = soup.get_text()
#break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
#break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
#drop blank lines
    text = ' '.join(chunk for chunk in chunks if chunk)
    df_red.at[i, 'description']=text.replace('ё', 'е').lower()
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords

ru_stemmer = SnowballStemmer("russian")
analyzer = TfidfVectorizer().build_analyzer()

def stemmed_words(doc):
    return (ru_stemmer.stem(w) for w in analyzer(doc))

stop_words_ru=stopwords.words('english')+stopwords.words('russian')+areas_tokens+['ст', 'м', 'г', 'пр', 'кт', 'шоссе', 'ул','московская', 'обл', 'район', 'по', 'для', 'со', 'от', 'на', 'без', 'of', 'and', 'on']
#print(stop_words_ru)

def stem_text(text):
    word_tokens = word_tokenize(text)
    stopped_text = [word for word in word_tokens if word not in stop_words_ru]
    stemmed_doc = [ru_stemmer.stem(doc) for doc in stopped_text]
    return stemmed_doc
```

```python
stemmed_names = [stem_text(doc) for doc in names_text]
stemmed_names_joined = [' '.join(l) for l in stemmed_names]
name_vectorizer = TfidfVectorizer(min_df=5, ngram_range=(1, 2))
names_bow = name_vectorizer.fit_transform(stemmed_names_joined)

stemmed_employers = [stem_text(doc) for doc in employers_text]
stemmed_employers_joined = [' '.join(l) for l in stemmed_employers]
employers_vectorizer = TfidfVectorizer(min_df=5, ngram_range=(1, 2))
employers_bow = employers_vectorizer.fit_transform(stemmed_employers_joined)
```

```python
print(len(name_vectorizer.vocabulary_)) 
print(len(employers_vectorizer.vocabulary_)) 
```

```python
stemmed_descriptions = [stem_text(doc) for doc in descriptions_text]
stemmed_descriptions_joined = [' '.join(l) for l in stemmed_descriptions]
descriptions_vectorizer = TfidfVectorizer(min_df=0.01, max_df=0.4, ngram_range=(1, 2)) # встречается в x to y% описаний
descriptions_bow = descriptions_vectorizer.fit_transform(stemmed_descriptions_joined)
```

```python
print(len(descriptions_vectorizer.vocabulary_))
```

```python
dum_n = pd.DataFrame(names_bow.toarray())
dum_e = pd.DataFrame(employers_bow.toarray())
dum_d = pd.DataFrame(descriptions_bow.toarray())

for k, v in name_vectorizer.vocabulary_.items():
    dum_n.rename(columns={v:'name_'+str(k)}, inplace=True)
    
for k, v in employers_vectorizer.vocabulary_.items():
    dum_e.rename(columns={v:'employer_'+str(k)}, inplace=True)

for k, v in descriptions_vectorizer.vocabulary_.items():
    dum_d.rename(columns={v:'description_'+str(k)}, inplace=True)
```
