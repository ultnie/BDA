```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(df_one_hoted.drop(['salary_to'], axis=1), df_y_train.salary_to, test_size=0.25)
```

```python
model_linear = LinearRegression()
model_linear.fit(X_train, y_train.values.ravel())

y_pred_train_lin = model_linear.predict(X_train)
y_pred_test_lin = model_linear.predict(X_test)

print(f'MAE train: {mean_absolute_error(y_train.values.ravel(), y_pred_train_lin)}, MAE test: {mean_absolute_error(y_test.values.ravel(), y_pred_test_lin)}')
print(f'MSE train: {mean_squared_error(y_train.values.ravel(), y_pred_train_lin)**0.5}, MSE test: {mean_squared_error(y_test.values.ravel(), y_pred_test_lin)**0.5}')
print(f'R2_score train: {r2_score(y_train.values.ravel(), y_pred_train_lin)}, R2 test: {r2_score(y_test.values.ravel(), y_pred_test_lin)}')
```

```python
pd.DataFrame(model_linear.coef_, index=X_train.columns).style.bar(align='mid', color=['#d65f5f', '#5fba7d'])
```

```python
from sklearn.linear_model import HuberRegressor

model_huber = HuberRegressor(max_iter=2000)
model_huber.fit(X_train, y_train.values.ravel())

y_pred_train_huber = model_huber.predict(X_train)
y_pred_test_huber = model_huber.predict(X_test)

print(f'MAE train: {mean_absolute_error(y_train.values.ravel(), y_pred_train_huber)}, MAE test: {mean_absolute_error(y_test.values.ravel(), y_pred_test_huber)}')
print(f'MSE train: {mean_squared_error(y_train.values.ravel(), y_pred_train_huber)**0.5}, MSE test: {mean_squared_error(y_test.values.ravel(), y_pred_test_huber)**0.5}')
print(f'R2_score train: {r2_score(y_train.values.ravel(), y_pred_train_huber)}, R2 test: {r2_score(y_test.values.ravel(), y_pred_test_huber)}')
```

```python
pd.DataFrame(model_huber.coef_, index=X_train.columns).style.bar(align='mid', color=['#d65f5f', '#5fba7d'])
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import precision_score, recall_score, f1_score

tree_model = DecisionTreeClassifier(max_depth=19, max_features='sqrt', criterion='entropy')
X_train_ordinal, X_test_ordinal, y_train_ordinal, y_test_ordinal = train_test_split(df_ordinal.drop(['salary_to'], axis=1), df_y_train.salary_to, test_size=0.25)
tree_model.fit(X_train_ordinal, y_train_ordinal.values.ravel())

y_pred_train_tree = tree_model.predict(X_train_ordinal)
y_pred_test_tree = tree_model.predict(X_test_ordinal)

print(f'Precision train: {precision_score(y_train_ordinal.values.ravel(), y_pred_train_tree, average="micro")}, Precision test: {precision_score(y_test_ordinal.values.ravel(), y_pred_test_tree, average="micro")}')
print(f'Recall train: {recall_score(y_train_ordinal.values.ravel(), y_pred_train_tree, average="micro")**0.5}, Recall test: {recall_score(y_test_ordinal.values.ravel(), y_pred_test_tree, average="micro")**0.5}')
print(f'F1 train: {f1_score(y_train_ordinal.values.ravel(), y_pred_train_tree, average="micro")}, F1 test: {f1_score(y_test_ordinal.values.ravel(), y_pred_test_tree, average="micro")}')
```

```python
#one hotting
from numpy.core.defchararray import add

ind_of_prof = sn[sn<.90].index.values
len(np.unique(ind_of_prof))
ind_of_emp = nen[nen<.35].index.values
len(np.unique(ind_of_emp))

one_hot_col = ['short_name', 'short_employer_name']

df_x_test.loc[~df_x_test['short_name'].isin(ind_of_prof), 'short_name'] = 'Rare'
df_x_test.loc[~df_x_test['short_employer_name'].isin(ind_of_emp), 'short_employer_name'] = 'Rare'

np_emp = np.append(ind_of_emp, 'Rare')
np_prof = np.append(ind_of_prof, 'Rare')

np_emp.sort()
np_prof.sort()

np_emp = 'short_employer_name_' + np_emp
np_prof = 'short_name_' + np_prof
df_x_test_dummies = pd.get_dummies(df_x_test[one_hot_col], prefix=one_hot_col, dtype=bool)
display(df_x_test_dummies)
df_x_test_dummies = df_x_test_dummies.reindex(columns = np.concatenate((np_prof, np_emp)), fill_value=False)
display(df_x_test_dummies)

df_x_test_one_hoted = pd.concat([df_x_test_dummies, df_x_test.select_dtypes('number'), df_x_test.select_dtypes('bool')], axis=1)

df_x_test_one_hoted
```
